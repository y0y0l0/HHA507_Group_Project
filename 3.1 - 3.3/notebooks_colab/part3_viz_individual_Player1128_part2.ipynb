T-test to determine relativity of metrics

````r
library(dplyr)


player_data <- read.csv('/content/player_1128_data.csv')

player_data$timestamp <- as.Date(player_data$timestamp, format = "%m/%d/%Y")


merged_player_data <- player_data %>%
  group_by(timestamp) %>%
  summarize(
    accel_load_accum = first(na.omit(accel_load_accum)),
    max_force_difference = first(na.omit(max_force_difference)),
    .groups = 'drop'
  )


cat("Head of merged data:\n")
print(head(merged_player_data))


cat("\nOriginal dimensions:", nrow(player_data), "rows,", ncol(player_data), "columns\n")
cat("Merged dimensions:", nrow(merged_player_data), "rows,", ncol(merged_player_data), "columns\n")
````

````r
library(dplyr)


cat("Column types of merged_player_data:\n")
print(sapply(merged_player_data, class))

non_na_accel_load_accum <- sum(!is.na(merged_player_data$accel_load_accum))
cat("\nNumber of non-NA values for 'accel_load_accum': ", non_na_accel_load_accum, "\n")


non_na_max_force_difference <- sum(!is.na(merged_player_data$max_force_difference))
cat("Number of non-NA values for 'max_force_difference': ", non_na_max_force_difference, "\n")
````

````r
library(dplyr)

t_test_results <- t.test(merged_player_data$accel_load_accum, merged_player_data$max_force_difference)

# Print the t-test results
cat("\nIndependent Samples t-test Results:\n")
print(t_test_results)
````

Does the player increase or decrease in performance?

````r
library(dplyr)

player_data_trends <- read.csv('/content/player_1128_data.csv')

player_data_trends$timestamp <- as.Date(player_data_trends$timestamp, format = "%m/%d/%Y")


earliest_timestamp <- min(player_data_trends$timestamp, na.rm = TRUE)


player_data_trends$time_since_start <- as.numeric(difftime(player_data_trends$timestamp, earliest_timestamp, units = "days"))

df_distance_total_clean <- player_data_trends %>%
  filter(!is.na(distance_total))

df_accel_load_accum_clean <- player_data_trends %>%
  filter(!is.na(accel_load_accum))

cat("\nHead of df_distance_total_clean:\n")
print(head(df_distance_total_clean))
cat("\nDimensions of df_distance_total_clean:", nrow(df_distance_total_clean), "rows,", ncol(df_distance_total_clean), "columns\n")

cat("\nHead of df_accel_load_accum_clean:\n")
print(head(df_accel_load_accum_clean))
cat("\nDimensions of df_accel_load_accum_clean:", nrow(df_accel_load_accum_clean), "rows,", ncol(df_accel_load_accum_clean), "columns\n")
````

````r
library(ggplot2)

regression_distance_total <- lm(distance_total ~ time_since_start, data = df_distance_total_clean)

regression_accel_load_accum <- lm(accel_load_accum ~ time_since_start, data = df_accel_load_accum_clean)

cat("\nSummary for Linear Regression of Distance Total:\n")
print(summary(regression_distance_total))

cat("\nSummary for Linear Regression of Accel Load Accum:\n")
print(summary(regression_accel_load_accum))
````

Does player show increase or decrease in performance

````r
library(ggplot2)

plot_distance_total <- ggplot(df_distance_total_clean, aes(x = time_since_start, y = distance_total)) +
  geom_point(alpha = 0.6, color = "blue") + # Add points for actual data
  geom_line(color = "darkblue", size = 0.8) + # Add lines connecting data points
  geom_smooth(method = "lm", se = FALSE, color = "red", linetype = "dashed") + # Overlay regression line
  labs(
    title = "Distance Total Over Time for Player 1128",
    x = "Days Since First Record",
    y = "Distance Total"
  ) +
  theme_minimal()

print(plot_distance_total)

plot_accel_load_accum <- ggplot(df_accel_load_accum_clean, aes(x = time_since_start, y = accel_load_accum)) +
  geom_point(alpha = 0.6, color = "darkgreen") + # Add points for actual data
  geom_line(color = "forestgreen", size = 0.8) + # Add lines connecting data points
  geom_smooth(method = "lm", se = FALSE, color = "purple", linetype = "dashed") + # Overlay regression line
  labs(
    title = "Acceleration Load Accumulation Over Time for Player 1128",
    x = "Days Since First Record",
    y = "Acceleration Load Accumulation"
  ) +
  theme_minimal()

print(plot_accel_load_accum)
````

Note : Player's performance is decreased overtime based on accel load accumulation and distance total metrics.

````r
library(dplyr)


df_max_force_difference_clean <- player_data_trends %>%
  filter(!is.na(max_force_difference))

df_torque_difference_clean <- player_data_trends %>%
  filter(!is.na(torque_difference))

cat("\nHead of df_max_force_difference_clean:\n")
print(head(df_max_force_difference_clean))
cat("\nDimensions of df_max_force_difference_clean:", nrow(df_max_force_difference_clean), "rows,", ncol(df_max_force_difference_clean), "columns\n")

cat("\nHead of df_torque_difference_clean:\n")
print(head(df_torque_difference_clean))
cat("\nDimensions of df_torque_difference_clean:", nrow(df_torque_difference_clean), "rows,", ncol(df_torque_difference_clean), "columns\n")
````

````r
#Linear regression

library(ggplot2)


regression_max_force_difference <- lm(max_force_difference ~ time_since_start, data = df_max_force_difference_clean)


regression_torque_difference <- lm(torque_difference ~ time_since_start, data = df_torque_difference_clean)

# Display summaries of the regression models
cat("\nSummary for Linear Regression of Max Force Difference:\n")
print(summary(regression_max_force_difference))

cat("\nSummary for Linear Regression of Torque Difference:\n")
print(summary(regression_torque_difference))
````

````r
library(ggplot2)

plot_max_force_difference <- ggplot(df_max_force_difference_clean, aes(x = time_since_start, y = max_force_difference)) +
  geom_point(alpha = 0.6, color = "darkred") + # Add points for actual data
  geom_line(color = "firebrick", size = 0.8) + # Add lines connecting data points
  geom_smooth(method = "lm", se = FALSE, color = "darkorange", linetype = "dashed") + # Overlay regression line
  labs(
    title = "Max Force Difference Over Time for Player 1128",
    x = "Days Since First Record",
    y = "Max Force Difference"
  ) +
  theme_minimal()

print(plot_max_force_difference)

plot_torque_difference <- ggplot(df_torque_difference_clean, aes(x = time_since_start, y = torque_difference)) +
  geom_point(alpha = 0.6, color = "darkgreen") + # Add points for actual data
  geom_line(color = "forestgreen", size = 0.8) + # Add lines connecting data points
  geom_smooth(method = "lm", se = FALSE, color = "purple", linetype = "dashed") + # Overlay regression line
  labs(
    title = "Torque Difference Over Time for Player 1128",
    x = "Days Since First Record",
    y = "Torque Difference"
  ) +
  theme_minimal()

print(plot_torque_difference)
````

Trends seen on visualizations do correlate with findings from literature review. With this particular player, there is an inverse relationship between torque-difference and max force difference versus distance total and acceleration accumulation load.
