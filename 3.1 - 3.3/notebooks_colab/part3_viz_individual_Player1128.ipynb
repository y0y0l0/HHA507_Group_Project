```r
print("Structure of player1128_mf_diff:")
str(player1128_mf_diff)

# Check for NA counts in the relevant columns
na_counts <- colSums(is.na(player1128_mf_diff[, c(
    "timestamp_datetime",
    "accel_load_accum",
    "distance_total",
    "max_force_difference",
    "torque_difference"
)]))

print("NA counts for relevant columns in player1128_mf_diff:")
print(na_counts)
```

```r
model1_data <- na.omit(player1128_mf_diff[, c(
    "timestamp_datetime",
    "distance_total",
    "max_force_difference",
    "torque_difference"
)])

print("First few rows of data for Model 1 (distance_total):")
print(head(model1_data))
print("Dimensions of data for Model 1:")
print(dim(model1_data))
```

```r
model_distance_data <- player1128_mf_diff[, c("timestamp_datetime", "distance_total")]


model_distance_data <- na.omit(model_distance_data)


model_distance_data$timestamp_numeric <- as.numeric(model_distance_data$timestamp_datetime)


print("First few rows of model_distance_data:")
print(head(model_distance_data))
print("Dimensions of model_distance_data:")
print(dim(model_distance_data))
```

```r
model_distance_total_simple <- lm(distance_total ~ timestamp_numeric, data = model_distance_data)


print("Summary for Simple Regression Model: distance_total vs. timestamp_numeric")
print(summary(model_distance_total_simple))


print("ANOVA Table for Simple Regression Model: distance_total vs. timestamp_numeric")
print(anova(model_distance_total_simple))
```
