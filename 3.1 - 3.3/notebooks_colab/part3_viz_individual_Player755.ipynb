```r
model_distance_data <- player755_mf_diff[, c("timestamp_datetime", "distance_total")]

# Remove any rows with NA values in these columns
model_distance_data <- na.omit(model_distance_data)

# Convert timestamp_datetime to a numeric format (Unix timestamp)
model_distance_data$timestamp_numeric <- as.numeric(model_distance_data$timestamp_datetime)

# Display the first few rows and dimensions of the prepared data
print("First few rows of model_distance_data:")
print(head(model_distance_data))
print("Dimensions of model_distance_data:")
print(dim(model_distance_data))
```

```r
model_distance_total_simple <- lm(distance_total ~ timestamp_numeric, data = model_distance_data)

# Summary of the regression model
print("Summary for Simple Regression Model: distance_total vs. timestamp_numeric")
print(summary(model_distance_total_simple))

# ANOVA for the regression model
print("ANOVA Table for Simple Regression Model: distance_total vs. timestamp_numeric")
print(anova(model_distance_total_simple))
```
