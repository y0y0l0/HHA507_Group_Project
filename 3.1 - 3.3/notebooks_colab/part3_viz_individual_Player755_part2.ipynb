T-test

````r
t_test_result <- t.test(
  player755_mf_diff$distance_total,
  player755_mf_diff$torque_difference,
  paired = FALSE
)

# Print the results of the t-test
cat("\nPaired t-test results for distance_total vs. torque_difference:\n")
print(t_test_result)

# Extract and print key findings
cat("\nInterpretation:\n")
cat(paste0("  Mean of differences (distance_total - torque_difference): ", round(t_test_result$estimate, 3), "\n"))
cat(paste0("  p-value: ", round(t_test_result$p.value, 4), "\n"))
cat(paste0("  95% Confidence Interval for the difference: [", round(t_test_result$conf.int[1], 3), ", ", round(t_test_result$conf.int[2], 3), "]\n"))

if (t_test_result$p.value < 0.05) {
  cat("  Since the p-value is less than 0.05, we reject the null hypothesis.\n")
  cat("  There is a statistically significant difference between the mean distance_total and mean torque_difference.\n")
} else {
  cat("  Since the p-value is greater than or equal to 0.05, we fail to reject the null hypothesis.\n")
  cat("  There is no statistically significant difference between the mean distance_total and mean torque_difference.\n")
}
````

````r
library(ggplot2)
library(dplyr)

# Load the simple_player755_data
simple_player755_data <- read.csv('/content/player_755_data.csv', colClasses = c(timestamp = "character"))

# Convert timestamp to POSIXct using the correct format YYYY-MM-DD HH:MM:SS
simple_player755_data$timestamp <- as.POSIXct(simple_player755_data$timestamp, format = "%Y-%m-%d %H:%M:%S")

# Filter out rows where timestamp is NA BEFORE plotting
plot_data_final <- simple_player755_data[!is.na(simple_player755_data$timestamp), ]

# Check if plot_data_final is empty after filtering
if (nrow(plot_data_final) == 0) {
  cat("Warning: plot_data_final is empty after timestamp conversion and NA filtering. Please check timestamp format or data content.\n")
} else {
  # Sort the data by timestamp to ensure geom_line draws correctly
  plot_data_final <- plot_data_final %>% arrange(timestamp)

  # --- Plot: Total Distance Over Time ---
  plot_distance_total <- ggplot(plot_data_final, aes(x = timestamp, y = distance_total)) +
    geom_line(color = "purple", alpha = 0.7) +
    geom_smooth(method = "lm", se = FALSE, color = "red", linetype = "dashed") +
    labs(
      title = "Player 755: Total Distance Over Time",
      x = "Date",
      y = "Total Distance"
    ) +
    theme_minimal() +
    theme(axis.text.x = element_text(angle = 45, hjust = 1))

  print(plot_distance_total)
}
````
